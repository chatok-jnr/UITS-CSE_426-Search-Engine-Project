# **7** - Missing Value Handling

```python
# Define searchable and metadata column groups
SEARCHABLE_COLS = ["Title", "Short Intro", "Category", "Sub-Category", "Skills", "Instructors"]
METADATA_COLS   = ["URL", "Rating", "Duration", "Course Type", "Language"]

# Keep only columns that actually exist in the cleaned dataframe
SEARCHABLE_COLS = [c for c in SEARCHABLE_COLS if c in df.columns]
METADATA_COLS   = [c for c in METADATA_COLS   if c in df.columns]

print(f"Searchable columns  : {SEARCHABLE_COLS}")
print(f"Metadata columns    : {METADATA_COLS}")

# Fill missing values in searchable columns with empty string
for col in SEARCHABLE_COLS:
    df[col] = df[col].fillna("")

# Fill missing values in metadata columns with a readable placeholder
for col in METADATA_COLS:
    df[col] = df[col].fillna("N/A")

# Reset index so document IDs are clean consecutive integers starting from 0
df.reset_index(drop=True, inplace=True)

print(f"\nMissing values after handling:")
print(df[SEARCHABLE_COLS + METADATA_COLS].isnull().sum())
print(f"\nTotal documents: {len(df)}")
```

# **8** - Build Search Documents

```python
# Concatenate all searchable fields into a single 'search_document' field per row.
# A space separator ensures tokens from different fields do not accidentally merge.

def build_search_document(row, cols):
    """Concatenate specified columns into one string, skipping empty values."""
    parts = [str(row[col]).strip() for col in cols if str(row[col]).strip()]
    return " ".join(parts)

df["search_document"] = df.apply(
    lambda row: build_search_document(row, SEARCHABLE_COLS), axis=1
)

print("Search document field created.")
print(f"\nSample search documents:")
for i in range(min(3, len(df))):
    print(f"\n--- Document {i} ---")
    print(df.loc[i, "search_document"][:300])
```

# **10 — Text Preprocessing Functions**

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# ── Global objects (created once, reused everywhere) ─────────────────────────
STOP_WORDS = set(stopwords.words("english"))
STEMMER    = PorterStemmer()

def lowercase(text: str) -> str:
    """Convert text to lowercase."""
    return text.lower()

def remove_punctuation(text: str) -> str:
    """Remove all punctuation characters from text."""
    return re.sub(r"[^\w\s]", " ", text)

def tokenize(text: str) -> list:
    """Split text on whitespace into a list of tokens."""
    return text.split()

def remove_stopwords(tokens: list) -> list:
    """Remove common English stopwords from token list."""
    return [t for t in tokens if t not in STOP_WORDS]

def stem_tokens(tokens: list) -> list:
    """Apply Porter stemming to every token."""
    return [STEMMER.stem(t) for t in tokens]

def preprocess(text: str, apply_stemming: bool = True) -> list:
    """
    Full preprocessing pipeline:
      lowercase → remove punctuation → tokenize → remove stopwords → stem
    Returns a list of processed tokens.
    """
    if not isinstance(text, str) or not text.strip():
        return []
    text   = lowercase(text)
    text   = remove_punctuation(text)
    tokens = tokenize(text)
    tokens = remove_stopwords(tokens)
    if apply_stemming:
        tokens = stem_tokens(tokens)
    return tokens

# Quick sanity check
sample = "Introduction to Machine Learning with Python and Pandas!"
print(f"Input  : {sample}")
print(f"Output : {preprocess(sample)}")
```

# **11 — Apply Preprocessing**

```python
# Apply the preprocessing pipeline to every search document.
# Store the result as a list of tokens in a new column.

df["tokens"] = df["search_document"].apply(preprocess)

print("Preprocessing complete.")
print(f"\nSample token lists:")
for i in range(min(3, len(df))):
    print(f"\nDoc {i} | Title: {df.loc[i, 'Title'][:60]}")
    print(f"Tokens : {df.loc[i, 'tokens'][:20]} ...")

print(f"\nTotal documents processed: {len(df)}")
```

# **12 — Build Inverted Index**

```python
# Inverted Index structure:
#   index[term] = { doc_id: term_frequency, ... }
#
# doc_id is the DataFrame row index (integer).

inverted_index = defaultdict(lambda: defaultdict(int))

for doc_id, tokens in enumerate(df["tokens"]):
    for token in tokens:
        inverted_index[token][doc_id] += 1

# Convert inner defaultdicts to plain dicts for cleaner access
inverted_index = {term: dict(postings) for term, postings in inverted_index.items()}

print("Inverted index built successfully.")
print(f"\nSample postings (first 5 terms):")
for term in list(inverted_index.keys())[:5]:
    print(f"  '{term}' → {dict(list(inverted_index[term].items())[:5])}")
```

# **13 — Display Inverted Index Statistics**

```python
total_docs    = len(df)
unique_terms  = len(inverted_index)
total_postings = sum(len(p) for p in inverted_index.values())

print("=== Inverted Index Statistics ===")
print(f"Total documents    : {total_docs}")
print(f"Unique terms       : {unique_terms}")
print(f"Total postings     : {total_postings}")
print(f"Avg postings/term  : {total_postings / unique_terms:.2f}")

# Top 20 most frequent terms by document frequency (df = # docs containing the term)
term_doc_freq = {term: len(postings) for term, postings in inverted_index.items()}
top_terms = sorted(term_doc_freq.items(), key=lambda x: x[1], reverse=True)[:20]

print(f"\nTop 20 terms by document frequency:")
for rank, (term, df_count) in enumerate(top_terms, 1):
    print(f"  {rank:>2}. '{term}' — appears in {df_count} documents")
```

# **14 — Build TF Statistics**

```python
# TF(term, doc) = frequency of term in doc / total tokens in doc
# Stored as: tf_scores[doc_id][term] = tf_value

tf_scores = {}

for doc_id, tokens in enumerate(df["tokens"]):
    total_tokens = len(tokens)
    if total_tokens == 0:
        tf_scores[doc_id] = {}
        continue
    term_counts = defaultdict(int)
    for token in tokens:
        term_counts[token] += 1
    tf_scores[doc_id] = {
        term: count / total_tokens
        for term, count in term_counts.items()
    }

print("TF scores computed.")
print(f"\nSample TF values for document 0:")
sample_tf = list(tf_scores[0].items())[:10]
for term, tf in sample_tf:
    print(f"  '{term}': {tf:.5f}")
```

# **15 — Build IDF Statistics**

```python
# IDF(term) = log((1 + N) / (1 + df_t)) + 1
# This is smooth IDF — avoids division by zero and never produces negative scores.
# N   = total number of documents
# df_t = number of documents containing the term

N = len(df)

idf_scores = {}

for term, postings in inverted_index.items():
    df_t = len(postings)                          # document frequency of term
    idf_scores[term] = math.log((1 + N) / (1 + df_t)) + 1

print("IDF scores computed.")
print(f"\nSample IDF values:")
sample_terms = list(idf_scores.items())[:10]
for term, idf in sample_terms:
    print(f"  '{term}': {idf:.5f}")

# Show IDF range
idf_values = list(idf_scores.values())
print(f"\nIDF range  : min={min(idf_values):.4f}  max={max(idf_values):.4f}")
print(f"IDF mean   : {sum(idf_values)/len(idf_values):.4f}")
```

# **16 — Build TF-IDF Ranking Engine**

```python
# TF-IDF(term, doc) = TF(term, doc) * IDF(term)
#
# For a query with multiple terms, the document score is the
# SUM of TF-IDF scores for each query term found in that document.

def compute_tfidf_score(doc_id: int, query_terms: list) -> float:
    """
    Compute the TF-IDF score for a single document given a list of query terms.
    Returns the sum of TF-IDF values across all query terms.
    """
    score = 0.0
    doc_tf = tf_scores.get(doc_id, {})
    for term in query_terms:
        tf  = doc_tf.get(term, 0.0)
        idf = idf_scores.get(term, 0.0)
        score += tf * idf
    return score

def rank_documents(candidate_doc_ids: set, query_terms: list) -> list:
    """
    Given a set of candidate document IDs and preprocessed query terms,
    return a list of (doc_id, score) tuples sorted by score descending.
    """
    scored = [
        (doc_id, compute_tfidf_score(doc_id, query_terms))
        for doc_id in candidate_doc_ids
    ]
    scored.sort(key=lambda x: x[1], reverse=True)
    return scored

print("TF-IDF ranking engine ready.")
```

# **17 — Boolean Retrieval Functions**

```python
def get_posting_set(term: str) -> set:
    """Return the set of document IDs that contain the given (stemmed) term."""
    stemmed = STEMMER.stem(term.lower())
    postings = inverted_index.get(stemmed, {})
    return set(postings.keys())

def boolean_and(terms: list) -> set:
    """
    AND retrieval: return doc IDs that contain ALL given terms.
    Starts with the smallest posting list for efficiency.
    """
    if not terms:
        return set()
    posting_sets = sorted(
        [get_posting_set(t) for t in terms],
        key=len
    )
    result = posting_sets[0]
    for s in posting_sets[1:]:
        result = result & s
        if not result:          # early termination
            break
    return result

def boolean_or(terms: list) -> set:
    """OR retrieval: return doc IDs that contain ANY of the given terms."""
    if not terms:
        return set()
    result = set()
    for t in terms:
        result = result | get_posting_set(t)
    return result

# Quick test
print("Boolean AND test ('python', 'data'):")
print(f"  Matching doc IDs: {list(boolean_and(['python', 'data']))[:10]}")

print("\nBoolean OR test ('python', 'java'):")
print(f"  Matching doc IDs: {list(boolean_or(['python', 'java']))[:10]}")
```

# **18 — Snippet Generation**

```python
def generate_snippet(doc_id: int, query_terms: list, snippet_length: int = 200) -> str:
    """
    Extract a relevant snippet from the original (unprocessed) search document.

    Strategy:
      1. Find the first sentence or window of text that contains a query term.
      2. If no match found, return the first `snippet_length` characters.

    Parameters
    ----------
    doc_id        : integer row index of the document
    query_terms   : list of preprocessed (stemmed) query terms
    snippet_length: approximate character length of the returned snippet
    """
    raw_text = df.loc[doc_id, "search_document"]

    # Split into sentences naively on period/newline
    sentences = re.split(r"(?<=[.!?])\s+|\n", raw_text)

    for sentence in sentences:
        sentence_tokens = preprocess(sentence)
        # Check if any query term appears in this sentence's tokens
        if any(qt in sentence_tokens for qt in query_terms):
            snippet = sentence.strip()
            if len(snippet) > snippet_length:
                snippet = snippet[:snippet_length].rsplit(" ", 1)[0] + " …"
            return snippet

    # Fallback: return the beginning of the document
    fallback = raw_text[:snippet_length].strip()
    if len(raw_text) > snippet_length:
        fallback = fallback.rsplit(" ", 1)[0] + " …"
    return fallback

print("Snippet generation function ready.")
print("\nSample snippet for document 0:")
sample_query_terms = preprocess("machine learning")
print(generate_snippet(0, sample_query_terms))
```

# **19 — Search Function**

```python
def parse_query(raw_query: str):
    """
    Parse a raw query string and return (mode, terms).

    Supported modes:
      'AND'   — query contains the keyword AND  (e.g. "python AND pandas")
      'OR'    — query contains the keyword OR   (e.g. "python OR java")
      'MULTI' — multiple words, no boolean keyword (treated as implicit OR ranked by TF-IDF)
      'SINGLE'— single word

    Returns
    -------
    mode  : str  — one of 'AND', 'OR', 'MULTI', 'SINGLE'
    terms : list — raw (unstemmed) terms for snippet matching,
                   stemmed separately inside retrieval functions
    """
    raw_query = raw_query.strip()

    if " AND " in raw_query.upper():
        parts = re.split(r"\bAND\b", raw_query, flags=re.IGNORECASE)
        terms = [p.strip() for p in parts if p.strip()]
        return "AND", terms

    if " OR " in raw_query.upper():
        parts = re.split(r"\bOR\b", raw_query, flags=re.IGNORECASE)
        terms = [p.strip() for p in parts if p.strip()]
        return "OR", terms

    tokens = raw_query.split()
    if len(tokens) > 1:
        return "MULTI", tokens

    return "SINGLE", tokens


def search(raw_query: str, top_k: int = 10) -> list:
    """
    Main search function.

    Accepts a raw query string.
    Returns a list of result dicts, ranked by TF-IDF score.

    Each result dict contains:
        rank, score, title, snippet, url, doc_id
    """
    mode, raw_terms = parse_query(raw_query)

    # Preprocess query terms (stem them to match the index)
    stemmed_terms = []
    for t in raw_terms:
        stemmed_terms.extend(preprocess(t))

    if not stemmed_terms:
        print("Query produced no searchable terms after preprocessing.")
        return []

    # ── Retrieval ─────────────────────────────────────────────────────────────
    if mode == "AND":
        candidate_ids = boolean_and(raw_terms)
    elif mode == "OR":
        candidate_ids = boolean_or(raw_terms)
    elif mode in ("MULTI", "SINGLE"):
        # Use OR of all terms — ranking by TF-IDF differentiates relevance
        candidate_ids = boolean_or(raw_terms)
    else:
        candidate_ids = set()

    if not candidate_ids:
        print(f"No documents found for query: '{raw_query}'")
        return []

    # ── Ranking ───────────────────────────────────────────────────────────────
    ranked = rank_documents(candidate_ids, stemmed_terms)
    ranked = [(doc_id, score) for doc_id, score in ranked if score >= 0.1]
    ranked = ranked[:top_k]



    # ── Build result list ─────────────────────────────────────────────────────
    results = []
    for rank, (doc_id, score) in enumerate(ranked, 1):
        title   = df.loc[doc_id, "Title"] if "Title" in df.columns else "N/A"
        url     = df.loc[doc_id, "URL"]   if "URL"   in df.columns else "N/A"
        snippet = generate_snippet(doc_id, stemmed_terms)
        results.append({
            "rank"   : rank,
            "score"  : score,
            "title"  : title,
            "snippet": snippet,
            "url"    : url,
            "doc_id" : doc_id,
        })

    return results

print("Search function ready.")
```

# **20 — Result Formatting**

```python
def display_results(results: list, query: str):
    """
    Pretty-print search results to the notebook output.
    """
    print(f"\n{'='*70}")
    print(f"  Query : \"{query}\"")
    print(f"  Found : {len(results)} result(s)")
    print(f"{'='*70}")

    if not results:
        print("  No results to display.")
        return

    for r in results:
        print(f"\n  Rank  : #{r['rank']}")
        print(f"  Score : {r['score']:.6f}")
        print(f"  Title : {r['title']}")
        print(f"  Snip  : {r['snippet']}")
        print(f"  URL   : {r['url']}")
        print(f"  {'-'*66}")

print("Result formatter ready.")
```
